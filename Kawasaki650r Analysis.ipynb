###  Engine Speed vs Brake Power
     
     Engine speed is rotation of the crankshaft measured in revolutions per minute or RPMs. Brake power, Wb, is the power output of the engine. Engine Speed and brake power are related through the formula
$$
  \dot W_b = 2\pi\tau N
$$

### Mean Piston Speed vs Bmep
    
    Mean piston speed, Up, is the speed at which the piston travels. It is related to the engine speed with 
$$    
    \tilde U_p = 2Ns. 
$$    

    Bmep or Brake mean effective pressure is the work done per unit volume of the engine. Bmep is related to the mean piston speed with 
$$    
    bmep = \frac{4 \dot W_b}{n_c A_p \overline{U_p}}  
$$

### Engine Speed vs Volumetric Efficency

    Volumetric efficiency, ev, is the mass of the air and fuel between the intake manifold and the cylinder divided by the density of the inlet air, displaced volume, and engine speed divided by 2. It relates to the engine speed through 
$$    
    ev = \frac{\dot m_{in}}{P_i V_d \frac{N}{2}} 
$$

```python
xValues = np.linspace(-180, 180, num=1000)
sin = np.sin(np.deg2rad(xValues)) 
cos = np.cos(np.deg2rad(xValues))
sqrt = np.sqrt
```

```python
expr = 1 + ((10.8-1)/2)*(1 - cos)+(1/(2*0.3937))*(1 - sqrt(1 - 0.3937**2 * sin**2))
```

```python
appx = 1 + ((10.8-1)/2)*(1 - cos)
```

```python
plt.plot(xValues, expr, color='blue', label='Exact Volume', alpha=0.8)
plt.plot(xValues, appx, '--', color='orange', label='Appx. Volume', alpha=1)
plt.title("Crank Angle vs Cylinder Volume")
plt.xlabel("Crank Angle (deg)")
plt.ylabel("Cylinder Volume (cubic cm)")
plt.legend()
plt.show()
```

Both the graphs appear to be exact but with closer examination the approximate appears to be a little lower than the exact in the six centimeter cubed mark. This graph also shows that the cylinder volume peaks at -180 degrees and 180 degrees.

```python
x = np.linspace(0, 180, num=1000)
Sin = np.sin(np.deg2rad(x))
Cos = np.cos(np.deg2rad(x))
Pi = np.pi
```

```python
pv = (Pi/2)*Sin*(1+ (0.3937*Cos)/sqrt(1-0.3937**2 * Sin**2))
```

The first half of the graph shows a steep slope while the second half of the graph the slope is not as steep. This graph also shows that the peak piston velosity is around 1.7 meters per second at around a 70 degree crank angle.
